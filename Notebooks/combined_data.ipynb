```python
import os
import glob
import pandas as pd

# 1Ô∏è‚É£ Input & output folders
INPUT_DIR = r"C:\Users\mh309\Downloads\Airbnb_data"       # folder with original CSV files
OUTPUT_DIR = r"C:\Users\mh309\Downloads\new_airbnb"

os.makedirs(OUTPUT_DIR, exist_ok=True)

# 2Ô∏è‚É£ City ‚Üí Country mapping
city_country_map = {
    "amsterdam": "Netherlands",
    "athens": "Greece",
    "barcelona": "Spain",
    "berlin": "Germany",
    "budapest": "Hungary",
    "lisbon": "Portugal",
    "london": "United Kingdom",
    "paris": "France",
    "rome": "Italy",
    "vienna": "Austria"
}
```

```python

# 3Ô∏è‚É£ Loop over files
for file_path in glob.glob(os.path.join(INPUT_DIR, "*.csv")):

    # Read CSV
    df = pd.read_csv(file_path)

    # Extract filename info
    filename = os.path.basename(file_path).replace(".csv", "")
    city, day_type = filename.split("_")

    city = city.lower()
    day_type = day_type.lower()

    # Add columns
    df["city"] = city
    df["day_type"] = day_type
    df["country"] = city_country_map.get(city, "Unknown")

    # Save new file
    new_filename = f"{city}_{day_type}_with_country.csv"
    output_path = os.path.join(OUTPUT_DIR, new_filename)

    df.to_csv(output_path, index=False)

    print(f"‚úÖ Saved: {new_filename}")

print("üéâ All files processed successfully!")
```
